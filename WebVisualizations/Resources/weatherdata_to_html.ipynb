```python
weather_data = 'cities.csv'
```

```python
weather_df = pd.read_csv(weather_data, encoding="ISO-8859-1")
```

```python
weather_df
```

```python
# inspect info of datatypes and possible null values
weather_df.info()
```

```python
# convert the epoch time column to datetime and format as Y-m-d
weather_df['Date_2'] = pd.to_datetime(weather_df['Date'], unit='s').dt.strftime('%Y-%m-%d')
weather_df
```

```python
# sort the data by new date column, in ascneding order
weather_df = weather_df.sort_values(by='Date_2', ascending=True)
weather_df
```

```python
# get list of columns in order to rearrange their order
cols = list(weather_df.columns.values)
cols
```

```python
weather_df = weather_df[['City_ID','City','Cloudiness','Country','Date','Date_2','Humidity','Lat','Lng','Max Temp','Wind Speed']]
weather_df
```
